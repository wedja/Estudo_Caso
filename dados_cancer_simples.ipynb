Importação de Bibliotecas



```python
import pandas as pd
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Carrega os Dados

```python
previsores = pd.read_csv('/content/drive/MyDrive/Deep_Learning/DataFrame/entradas_breast.csv') #Base de dados de entrada
classes = pd.read_csv('/content/drive/MyDrive/Deep_Learning/DataFrame/saidas_breast.csv') #Base de dados de saída
```

Separação dos dados para treinamento e teste



```python
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classes, test_size=0.25)
```

Arquitetura da rede neural

```python
classificador = Sequential()
classificador.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform', input_dim = 30))
classificador.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform'))
classificador.add(Dense(units=1, activation='sigmoid'))
classificador.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['binary_accuracy'])
```

Treinando o modelo

```python
classificador.fit(previsores_treinamento, classe_treinamento, batch_size=10, epochs=100)
```

```python
previsoes = classificador.predict(previsores_teste)
print(previsoes)
```

```python
previsoes = (previsoes > 0.5)
previsoes
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
precisao = accuracy_score(classe_teste, previsoes)
display(precisao)
```

```python
matriz = confusion_matrix(classe_teste, previsoes)
display(matriz)
```

```python
resultado = classificador.evaluate(previsores_teste, classe_teste)
print(resultado)
```
